```python
#Carregando os dados
import pandas as pd
```

```python
tabela=pd.read_csv('C:/Users/paulo/OneDrive/Área de Trabalho/Python Pandas/Kaggle/wine_dataset.csv')
tabela.head()
```

```python
#Trocando os valores do vinho por 0 e 1
tabela['style'] = tabela['style'].replace('red',0)
tabela['style'] = tabela['style'].replace('white',1)
```

```python
#Separando as variáveis
y = tabela['style']
x = tabela.drop('style', axis = 1)
```

```python
#Criando o Treino e Teste
from sklearn.model_selection import train_test_split
x_treino, x_test, y_treino, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
#Criando o modelo
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
#Imprimindo o resultado
resultado = modelo.score(x_test, y_test)
print('Probabilidade: ',resultado)
```
